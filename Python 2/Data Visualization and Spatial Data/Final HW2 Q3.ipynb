```python
country_names = ['Tanzania', 'Botswana']
Indicator = Africa['Indicator'].unique()
Year = Africa['year'].unique()

from matplotlib.colors import Normalize
from mpl_toolkits.axes_grid1 import make_axes_locatable




def plot_country(country_names, Indicator, Year):
    #function from above, but without the color generator
    plot = fig, ax = plt.subplots(1)
    #norm = Normalize(vmin = min(Africa[Africa['Indicator']== Indicator]['Variable']), vmax = max(Africa[Africa['Indicator']== Indicator]['Variable']))
    from mpl_toolkits.axes_grid1 import make_axes_locatable
    divider = make_axes_locatable(ax)
    cax = divider.append_axes('right', size='5%', pad=0.1)
    Africa.plot(ax=ax, column='Variable', edgecolor = 'black', legend = True, color = 'white')

    ax = Africa[Africa.name == country_names].plot(ax=ax, cax=cax, column='Variable', legend=True, lw=1, edgecolor = 'black')

   
    ax.axis('off');
    #return plot
    
@interact(name=country_names, Indicator = Indicator, Year =Year)

def make_plot_for(name=country_names, Indicator = Indicator, Year = Year):
    plot = plot_country( name , Indicator, Year)
```
